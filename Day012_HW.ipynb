### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")

for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all('a')[-1].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
news_list = {}
for d in soup.find(class_="part_list_2").find_all('h3'):
    if not news_list.get(d.find('em').text):
        news_list[d.find('em').text] = []
    news_list[d.find('em').text].append({'日期':d.find(class_="date").text,'標題':d.find_all('a')[-1].text})    
news_list
```

### ③ 哪一個類別的文章最多

```python
max = ['',0]
for key, value in news_list.items():
    if len(value) >= max[1]:
        max[0] = key
        max[1] = len(value)
print(max)
```
